```python
#Calcular ventas por Regin

ventas_por_region = df.groupby('Region')['Ventas'].sum()

print('A continuacin se presentan las Ventas por',ventas_por_region)
```

```python
#Filtrar el dataframe para mostrar solo las ventas realizadas por el vendedor "David"

ventas_david = df[df['Vendedor'] == 'David']
total_ventas_david = ventas_david['Ventas'].sum()

print('Las ventas de David en total son', total_ventas_david, '\n', '\n', 'A continuacin el detalle organizado por fecha:','\n',ventas_david)
```

```python
#Calcular el promedio de ventas por regin

ventas_por_producto = df.groupby('Region')['Ventas'].mean().astype(int)
print('El promedio de ventas por regin es el siguiente:','\n', '\n',ventas_por_producto)
```

```python
#Calcular el promedio de ventas por regin con dos decimales

ventas_por_producto = df.groupby('Region')['Ventas'].mean().round(2)
print('El promedio de ventas por regin es el siguiente:','\n', '\n',ventas_por_producto)
```

```python
#Ordenar el df por ventas en orden descendente

df_ordenado = df.sort_values(by='Ventas', ascending=False)
print('Dataframe organizadon de manera descendente desde la columna de Ventas:','\n',df_ordenado)
```

```python
#Filtrar las filas con ventas mayores a 10000 y ordenar de mayor a menor

df_filtrado =df[df['Ventas'] > 10000].sort_values(by='Ventas', ascending=False)
print('A continuacin se presenta las ventas mayores a 10000 organizadas de mayor a menor','\n',df_filtrado)
```

```python
#Filtrar las filas con ventas mayores a 10000 y ordenar de mayor a menor

df_filtrado =df[df['Ventas'] > 10000].sort_values(by='Ventas', ascending=False)
print('A continuacin se presenta las ventas mayores a 10000 organizadas de mayor a menor','\n',df_filtrado)
```
